```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
obv = pd.read_csv('../policies/SAC_2025-11-25_10-59-34/test_results_SAC_2025-11-25_10-59-34.csv')
# Box plot of each observation component
obv_columns = obv.columns.tolist()
plt.figure(figsize=(12, 6))
plt.boxplot([obv[col] for col in obv_columns], labels=obv_columns)
plt.title('Box Plot of Observation Components')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Print standard deviation and mean of each observation component
for col in obv_columns:
    print(f"{col}: mean = {obv[col].mean():.4f}, std = {obv[col].std():.4f}")
```
